```python
import sys
import logging
import numpy
import matplotlib
from matplotlib import pyplot
from mpl_toolkits.mplot3d import axes3d
from scipy.integrate import odeint, ode
from PIL import Image
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import maptools
```

```python
def map(xi, n_steps, r):
    """ Function doc """
    x = numpy.zeros(n_steps)
    x[0] = xi
    for i in range(n_steps-1):
        x[i+1] = r * x[i] * (1.0 - x[i])
        # x[i+1] = numpy.cos(r * x[i])
    return x

def run(y0, r_coeff = 3.0):
    n_steps = 20      # number of steps per map

    # Trajectory plot
    # y0 = 0.5            # initial value
    xt = numpy.linspace(0.0, 1.0, n_steps)
    yt = map(y0, n_steps, r_coeff)
    print([xt])
    print([yt])

    # Cobweb plot
    n_values = 10
    x = numpy.linspace(0.0, 1.0, n_values)
    y =  r_coeff * x * (1.0 - x)

    # Display
    maptools.display(x, y, xt, yt)

run(0.2, 2.6)

# interact(run, r=(2.5, 3.5))
```
